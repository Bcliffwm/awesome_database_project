```python
from google.cloud import bigquery
import os
```

```python
os.environ['GOOGLE_APPLICATION_CREDENTIALS'] = '/content/amazing-database-project-bf603890e975.json'
```

```python
client = bigquery.Client()
```

```python
sql_query = """
  SELECT r.id, r.title, r.rating,
  n.calories, n.carbs, n.cholesterol, n.fiber, n.protein, n.sodium, n.sugar
  FROM recipes.recipes_data AS r
  INNER JOIN recipes.nutrients_dataset AS n
  ON r.id = n.recipe_id
  ORDER BY r.id ASC
  LIMIT 5
 """
```

```python
query_job = client.query(sql_query)
```

```python
for row in query_job.result():
  print(row)
```
